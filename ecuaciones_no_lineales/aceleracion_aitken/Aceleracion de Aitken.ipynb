# Aceleración de Aitken

### **Teorema 1**

Sea $(r_{n})$ una sucesión numérica convergente al limite $r$. La nueva sucesión 
$$s_{n} = \frac{r_{n}r_{n+2}-r_{n+1}^{2}}{r_{n+2}-2r_{n+1}+r_{n}},\quad n\geq 0$$
converge a $r$ más rapido, si $r_{n+1}-r = (c+\delta_{n})(r_{n}-r)$ con $\vert c\vert <1$ y $\lim_{n\rightarrow\infty}\delta_{n}=0$.

```python
"""
Implementación de la aceleración de Aitken.
Gráfica de comparación de la rapidez de convergencia de (rn) y (sn)

r: sucesión que converge a 'r'
s: sucesion que resulta al aplicar la aceleración de Aitken a la sucesión r
"""


def aceleracion_aitken(r):
    return lambda n: (r(n)*r(n+2)-r(n+1)**2)/(r(n+2)-2*r(n+1)+r(n))
    
import numpy as np
import itertools as itts
import matplotlib.pyplot as plt

# comparacion la rapidez de convergencia de r y s
def rapidez_convergencia(r, MAX_ITERATIONS=10):
    n = np.arange(1, MAX_ITERATIONS)
    s = aceleracion_aitken(r)
    
    plt.plot(n, r(n), label='r(n)')
    plt.plot(n, s(n), label='s(n)')
    plt.grid()
    plt.title("Comparación de la rapidez de convergencia")
    plt.legend()
```

```python
"""
Problema de Aplicación.

Compare la rapidez de convergencia de la sucesión rn = sin(n)/n con la sucesion generada por la aceleración de 
Aitken
"""

r = lambda n: np.sin(n)/n
rapidez_convergencia(r)
```
